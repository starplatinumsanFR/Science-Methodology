For the first figure I will be criticizing, I chose the graph about Gun Deaths in Florida:

This figure targets the general audience of citizens in America, specifically those living in Florida as gun deaths are caused by civilians and they directly impact their lives. However, this graphic fails to effectively show the proportions of crimes committed by guns before and after the ''Stand your ground act'' was enacted.

First of all, the graph uses a filled plot are which exagerates the actual data and makes it seem worse than it is, probably to convey fear and reactions rather than the reality and how the new law changed the situation over time.

Not only is the Y-axis reversed, but they are using the upper side of it to color it in red to represent the effect of blood spilling from a wall towards the ground. A simple line plot would be more adapted to share the data without ambiguity. The red also introduces emotional bias. The filled area could definitely be removed without any information loss.

The graph also shows no explicit building method such as if it is by interpolation or discrete data.

There are no confidence intervals whatsoever shown in the graph, no histograms to vizualise the probability as well.

The X-axis is extremely wide as it only splits in between decades without individual years, making it hard to know what year each plot is from. It also makes each dot unclear what year it is from unless you count them out and assume they actually picked that pattern of stat calculations. The grid lines are therefore absent.

The units are implicitely indicated.

There's no way to compare this data with any other parameter like a different state, a different context etc.
